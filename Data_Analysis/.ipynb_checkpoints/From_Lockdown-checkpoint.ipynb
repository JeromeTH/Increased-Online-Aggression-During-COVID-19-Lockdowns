```python
def out_of_range(date, data):
    return date < tm.get_date(data.iloc[0]['start_date']) or date > tm.get_date(data.iloc[-1]['end_date'])

def get_week(date, data):
    l = 0
    r = 92
    while l < r:
        mm = l + (r - l)//2
        now = data.loc[mm] # now is a dictionary
        st = tm.get_date(now['start_date'])
        ed = tm.get_date(now['end_date'])
        if ed < date:
            l = mm + 1
        else:
            r = mm
        
    return l;
```

```python
for filename in os.listdir(f"{FILE_DIR}"):
    print(filename)
    state = filename[:-5]
    try: start_date = lc.lockdown_date(state)
    except: continue
    
    if lc.no_data(state):
        with open(f"{FILE_DIR}/{filename}", "r") as file:
            cur = json.load(file)
            for i in range(len(cur)):
                cur[i]['weeks_from_lockdown'] = -100
        
        with open(f"{FILE_DIR}/{filename}", "w") as file:
            json.dump(cur, file)
        continue

    end_date = lc.open_date(state)
    print(start_date)
    # weeks 0 means the date is 2019/1/1 which means the state didn't lockdown 
    print(lc.no_data(state))
    print(get_week(start_date, lockdown))
    start_week = get_week(start_date, lockdown)
    end_week = get_week(end_date, lockdown)
    print(start_week)
    with open(f"{FILE_DIR}/{filename}", "r") as file:
        cur = json.load(file)
        for i in range(len(cur)):
            
            if start_week == 0 or i > end_week:
                cur[i]['weeks_from_lockdown'] = -100
                continue
            cur[i]['weeks_from_lockdown'] = i - start_week
        
    with open(f"{FILE_DIR}/{filename}", "w") as file:
        json.dump(cur, file)
            
    
```
